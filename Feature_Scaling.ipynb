### **Feature Scaling**

```python
import pandas as pd

df = pd.read_csv('employee_promotion.csv')
df.head()
```

```python
df_copy = df.copy()
```

## **Feature Scaling**

```python
# 1. MinMax Scaler
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
# Imports scaling tools from sklearn.

minmax_scaler = MinMaxScaler()
# Creates a Min-Max scaler object (scales data between 0–1).

df_copy["Salary_Scaled_Normalization"] = minmax_scaler.fit_transform(df_copy[["Salary"]])
# Fits scaler on Salary column and transforms values into 0–1 range, storing results in a new column.
```

```python
# 2. Standard Scaler
```

```python
standard_scaler = StandardScaler()
# Creates a Standard Scaler object (scales data with mean = 0 and standard deviation = 1).

df_copy["Salary_Scaled_Standardization"] = standard_scaler.fit_transform(df_copy[["Salary"]])
# Fits on the Salary column and transforms values using (x − mean)/std, storing standardized values in a new column
```
